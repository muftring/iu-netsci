# Exploring The Normal Distribution

```python
import math
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Numpy [link](https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html)

```python
# Mean (“centre”) of the distribution.
mu = 0.0 

# Standard deviation (spread or “width”) of the distribution.
sigma = 0.01

# Output shape (number of samples to be drawn)
size = 1000
```

```python
# Drawn samples from the parameterized normal distribution.
dist = np.random.normal(mu, sigma, size)
```

```python
count, bins, ignored = plt.hist(dist, bins=30, normed=True)

plt.plot(bins, 
         1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, 
         color='r')

plt.show()
```

some constant values

```python
ITERATIONS = 300
MEAN = 4.0
STDDEV = 0.01
```

Keep the Standard Deviation fixed

```python
segs = 30
means = np.linspace(1.0, 6.0, segs)
sigma = STDDEV
size = 1000

ncols = 6
nrows = math.ceil(segs/ncols)
width = 16
height = 2 * nrows

plt.figure(1, [width, height])

plot_number = 0
for mu in means:
    dist = np.random.normal(mu, sigma, size)
    plot_number += 1
    plt.subplot(nrows, ncols, plot_number)
    count, bins, ignored = plt.hist(dist, bins=30, normed=True)
    plt.plot(bins,
             1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
             color='r')
plt.show()
```

keep the Mean fixed

```python
segs = 30
mu = MEAN
stddevs = np.linspace(0.01, 8.0, segs)
size = 1000

ncols = 6
nrows = math.ceil(segs/ncols)
width = 16
height = 2 * nrows

plt.figure(1, [width, height])

plot_number = 0
for sigma in stddevs:
    dist = np.random.normal(mu, sigma, size)
    plot_number += 1
    plt.subplot(nrows, ncols, plot_number)
    count, bins, ignored = plt.hist(dist, bins=30, normed=True)
    plt.plot(bins,
             1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
             color='r')
plt.show()
```
